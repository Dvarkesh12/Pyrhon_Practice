```python
# Splitwise Dataset Practice
import pandas as pd
## Importing csv file from the dataset
data = pd.read_csv('export.csv')
```

```python
##  .iloc[]: This is a method used to access rows and columns by integer position.
## [start row: end row: step size]
splitwise_data = data.iloc[1::]
# print('splitwise_data:',splitwise_data)

## How to find that two values are similar in two columns
DG_PP_similarexpense = splitwise_data[splitwise_data['Dvarkesh Ghatol']==splitwise_data['Pranjal patil']]
# print('DG_PP_similarexpense:',DG_PP_similarexpense)

## Find the unique values of expenses
Unique_Expenses = DG_PP_similarexpense['Dvarkesh Ghatol'].unique()
# print('Unique_Expenses:',Unique_Expenses)

## Sort the authors in ascending order
sorted_uniqueexpenses = sorted(Unique_Expenses)
# print('sorted_uniqueexpenses:',sorted_uniqueexpenses)


# Create a DataFrame with the sorted unique expense and rename the 'sorted_uniqueexpenses' column to 'rs'
paisa = pd.DataFrame({'rs': sorted_uniqueexpenses})
paisa
```
